```python
sex=pd.get_dummies(titanic['Sex'],drop_first=True) # we do this for getting sex colum to integers
```

```python
embarked=pd.get_dummies(titanic['Embarked'],drop_first=True)
```

```python
titanic=pd.concat([titanic,sex,embarked],axis=1) # add the integers values of colums in the data
```

```python
titanic.drop(['PassengerId','Name','Ticket','Sex','Embarked','Cabin'],axis=1,inplace=True)
```

 ## Creating the Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline=Pipeline([
    ('std_scaler', StandardScaler())
])
```
